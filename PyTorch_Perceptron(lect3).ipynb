```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from sklearn import datasets
import torch.nn as nn
```

```python
torch.manual_seed(1)
n_pts = 100
centers = [[-0.5,0.5],[0.5,-0.5]]
X,y = datasets.make_blobs(n_samples=n_pts,centers=centers,cluster_std=0.4,random_state=123)
x_data = torch.Tensor(X)
y_data = torch.Tensor(y.reshape(100,1))
```

```python
def scatter_plot():
  plt.scatter(X[y==0,0],X[y==0,1])
  plt.scatter(X[y==1,0],X[y==1,1])
```

```python
scatter_plot()
```

```python
class Model(nn.Module):
  def __init__(self,input_size,output_size):
    super().__init__()
    self.linear = nn.Linear(input_size,output_size)
  def forward(self,x):
    pred = torch.sigmoid(self.linear(x))
    return pred
  def predict(self, x):
      pred = self.forward(x)
      if pred >= 0.5:
        return 1
      else:
        return 0
```

```python
torch.manual_seed(2)
model=Model(2,1)
print(list(model.parameters()))
```

```python
[w,b] = model.parameters()
w1,w2 = w.view(2)
def get_params():
  return (w1.item(),w2.item(),b[0].item())
```

```python
def fit_plot(title):
  plt.title = title
  w1,w2,b1 = get_params()
  x1 = np.array([-2.0,2.0])
  x2 = (w1*x1+b1)/-w2
  plt.plot(x1,x2,"r")
  scatter_plot()
```

```python
fit_plot("Initial")
```

```python
criterion = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
epochs = 1000
losses = []

for i in range(epochs):
  y_pred = model.forward(x_data)
  loss = criterion(y_pred, y_data)
  print("epoch:", i, "loss:", loss.item())
  losses.append(loss.item())
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
plt.plot(range(epochs), losses)
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.grid()
```

```python
fit_plot("Trained Model")
```

```python
point1 = torch.Tensor([1.0, -1.0])
point2 = torch.Tensor([-1.0, 1.0])
plt.plot(point1.numpy()[0], point1.numpy()[1], 'ro')
plt.plot(point2.numpy()[0], point2.numpy()[1], 'ko')
fit_plot("Trained Model")
print("Red point positive probability = {}".format(model.forward(point1).item())) 
print("Black point positive probability = {}".format(model.forward(point2).item())) 
print("Red point belongs in class {}".format(model.predict(point1))) 
print("Black point belongs in class = {}".format(model.predict(point2)))
```
